### 3.2 Randomisation

```python
import random
import string
```

```python
def generate_random_string(length):
    letters = string.ascii_letters + string.digits
    return ''.join(random.choice(letters) for _ in range(length))
```

```python
num_unique_strings = 21747
```

```python
unique_strings = set()
while len(unique_strings) < num_unique_strings:
    unique_strings.add(generate_random_string(10))
```

```python
df2 = pd.DataFrame(list(unique_strings), columns=['RandomNameString'])

# Display the DataFrame
print(df2)
```
